```python
import pandas as pd
import numpy as np
```

Since the file contains multiple language characters, declare the encoding method with read_csv()

```python
# Read CSV file with encoding method 'utf-8' and tranfer it into dataframe
raw = pd.read_csv("../Data/winemag-data_first150k.csv",encoding='utf-8')
raw_df = pd.DataFrame(raw)
```

```python
raw_df.tail(10) # First 10 rows of the dataframe
```

With the dateframe has 150930 rows and 11 columns, we want to remove column "region_2" which contains a large amount of "NaN" data. Also rows that contains NaN values in the remaining dataframe.

```python
semi_df = raw_df.drop(['region_2'],axis=1).dropna()
semi_df.tail(10)
```

```python
semi_df['ID'] = np.arange(len(semi_df["ID"])) # Reset ID Column 
semi_df.tail(10)
```

```python
semi_df.to_csv('../Data/clean_winemag-data.csv',encoding = 'utf=8',index = False)
```
